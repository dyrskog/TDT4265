## task 1c)
Use the fact that $H_2 = W_2$ and $H_2 = H_1 = H$:

\begin{align*}
    H_2 &= \frac{H_1 - F + 2P}{S} + 1 \\
    H_2S &= H_1 - F + 2P + S \\
    H_2 &= H_1 = H \\
    2P &= H(S - 1) - S + F \\
    P &= \frac{H(S - 1) - S + F}{2} \\
    P &= \frac{H(1-1) - 1 + 7}{2} \\
    P &= 3
\end{align*}
